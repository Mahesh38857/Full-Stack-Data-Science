# Introduction to seaborn

## Seaborn is very powerfull visualizatio(STATISTIC VISULAIZATION) pkg in python


```python
import matplotlib.pyplot as plt # Visualization 
import seaborn as sns # Distraction Visualization
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
plt.rcParams['figure.figsize'] = 8,4
```

```python
stats.head()
```

```python
stats.head()
```

```python
vis1 = sns.distplot(stats["InternetUsers"])
vis1
plt.show()
```

```python
vis1 = sns.distplot(stats["InternetUsers"],bins = 10)
```

```python
vis1
plt.show()
```

```python
# Box plots:
vis2 = sns.boxplot(data = stats, x= "IncomeGroup", y='BirthRate',palette="tab10")
```

```python
vis2
plt.show()
```

```python
vis3 = sns.lmplot(data = stats,x = 'InternetUsers',y = 'BirthRate',fit_reg = False)
```

```python
vis3
plt.show()
```

```python
vis4 = sns.lmplot(data = stats,x = 'InternetUsers', y = 'BirthRate')
```

```python
vis4
plt.show()
```
