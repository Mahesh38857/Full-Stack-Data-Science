```python
import pathlib
import textwrap

import google.generativeai as genai

from IPython.display import display, Markdown  # Correct module and import

def to_markdown(text):
    text = text.replace('•', ' *')  # Converts bullet points to Markdown style
    return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```
